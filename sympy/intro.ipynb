```python
import sympy as sp

# Define symbols
x, y, z = sp.symbols('x y z')

# Display the symbols
print(x, y, z)
```

```python
# Use display and Math
from IPython.display import display, Math
from sympy import latex, symbols, diff

x, y = sp.symbols('x y')
expression = x**2 + 2*x*y + y**2
display(expression)

x = symbols('x')
y = diff(6*x**5)

display(Math(f' y = {latex(y)}'))
```

```python
# Function to solve an expression for a given variable
def solve_expr(expr, x) -> list:
  solution = solve(expr, x)
  return solution
```

```python
# When to use simplify vs factor
from sympy import *
import typing
# Define an expression
expr = x**2 + 2*x + 1
expr2 = x**2 + 2*x -5

# Simplify the expression
simplified = simplify(expr)
print(simplified)
display(simplified)

# Use factor to get the factored output
factored = factor(expr)
print(factored)

# Solve an equation
solution = solve_expr(expr, x)
print(f"Solution for expr {expr} is: {solution}")

solution2 = solve_expr(expr2,x)
print(f"Solution for expr {expr2} is: {solution2}")
```

```python
# Plot the above expression
from sympy.plotting import plot

# Plot the function sin(x)
plot(expr)
plot(expr2)
```
